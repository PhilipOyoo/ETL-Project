```python
import pandas as pd
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
df = pd.read_csv(r"C:\Users\phili\Desktop\ABnB_Data.csv")
```

```python
df.head()
```

```python
New_Data = df[['id','name','host_id','host_name','price','latitude','longitude']]
New_Data.set_index('id').head()
```

```python
New_Data["Long + Lat"] =  New_Data["latitude"].map(str) + "," +" " + New_Data["longitude"].map(str)
New_Data.head()

#  = dataframe["Year"].map(str) + dataframe["quarter"]
```

```python
New_Data.head()
```

```python
borough = []
loc = New_Data['Long + Lat']
for l in loc:
    sub = str(geolocator.reverse(l))
    borough.append(sub.split(', ')[2])
```
